```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Generate a list of random numbers
n = 10
arr = [random.randint(1, 100) for _ in range(n)]
print("Original List:", arr)

# Bubble Sort Function
def bubble_sort(arr):
    n_len = len(arr)
    for i in range(n_len - 1):
        for j in range(0, n_len - i - 1):
            # Visualize current state
            plt.bar(range(n_len), arr)
            plt.title(f"Bubble Sort: Pass {i+1}, Comparison {j+1}")
            plt.pause(0.01)
            plt.clf()

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap
    return arr

# Run Bubble Sort
sorted_arr = bubble_sort(arr)
plt.bar(range(n), sorted_arr)
plt.title("Bubble Sort: Final Sorted List")
plt.show()
```
